```
# 2 way - ADDING LEGEND TO MENTION WHICH LINE SHOW WHICH DATA.
import numpy as np

x_axis_value=np.arange(10,20,1)
y1_axis_value=np.arange(100,200,10)

y2_axis_value=np.arange(300,400,10)
plt.title("This is second plot")
plt.xlabel("Contain the x axis number.")
plt.ylabel("Contain the y axis value.")
plt.plot(x_axis_value,y1_axis_value,label="First line",color="red")
plt.plot(x_axis_value,y2_axis_value,label="Second data line plot",color="black")
#Legend
plt.legend()
#HERE INSIDE LEGEND I AM NOT PASSING THE ARGUMENT OF LABLE INSTEAD I AM PASSING IN PLT.PLOT ARGUMENT AS LABEL.
plt.show()
```

```
#BAR AND BAR BOTH IN SAME PLOT.


import numpy as np
from matplotlib import pyplot as plt 



plt.style.use('dark_background')
x_axis_value=np.arange(10,20,1)
y1_axis_value=np.arange(100,200,10)
y2_axis_value=np.arange(300,400,10)

x_index=np.arange(len(x_axis_value))# For x ticks.
width=0.3 # for giving gap. 

plt.title("This is second plot")
plt.xlabel("Contain the x axis number.")
plt.ylabel("Contain the y axis value.")

plt.grid(True)
plt.bar(x_index-width,y1_axis_value,color="red")
plt.bar(x_index+width,y2_axis_value,color="blue")

plt.xticks(ticks=x_index, labels=x_axis_value)####ADDED FOR XTICKS
plt.yticks(ticks=np.arange(0,600,50))

plt.show()
```
