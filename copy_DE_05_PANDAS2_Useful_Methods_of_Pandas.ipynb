```
##################################################
#User-Defined Functions with the apply Method

This method accepts a user-defined function and additional arguments and
returns a new column after applying the function on a particular column element-wise.

As an example, suppose we want to create a column of categorical features like high/
medium/low based on the sales price column. 
```

```
columnname.apply(functionname)

def userfun(value):
  return true


df['newcolumn']=df['column to analyze'].apply(userfun)
```

```
#create user define funciton

def caterize_sales_type(value):
  if value<50:
    return "Low"
  elif value<200:
    return "Medium"
  else:
    return "High"

    
df20=df[['Customer Name','State','Sales','Profit']].sample(n=20)
df20['using_apply']=df20['Sales'].apply(caterize_sales_type)
#IN above line i am creating one new column name - using_apply, inside this column
#i am inserting the return value of userdefine function.
print(df20)
```

```
#The apply method also works with the built-in native Python functions. 
#if suppose i want to findout the character present in customer Name.


df20['character is customer name:']=df20['Customer Name'].apply(len)
print(df20)
#print("         20    ")
#print("\t 20 \t")
```
